```python
import numpy as np
from matplotlib import pyplot as plt
from skimage import io, data

%matplotlib inline
```

```python
def power_iteration(A, iterations=1000):
    n_lambdas = A.shape[1]
    eig_vals = []
    eig_vecs = []
    for i in range(n_lambdas):
        vecs = np.eye(A.shape[0], A.shape[1])
        for j in eig_vecs:
            vecs -= np.dot(j,j.T)
        A = np.dot(A, vecs)
        
        x_k = np.random.rand(A.shape[0]).reshape(A.shape[0],1)
        for _ in range(iterations):
            x_k1 = np.dot(A, x_k)

            x_k1_norm = np.linalg.norm(x_k1)
            
            x_k = x_k1 / x_k1_norm
            
        eig_vecs.append(x_k)
        eig_vals.append(np.dot(x_k.T, np.dot(A, x_k))[0])

    return np.array(eig_vals), np.array(eig_vecs)
```

```python
A = np.diag([1,2])
A = data.coins()
w,v = np.linalg.eig(np.dot(A.T, A))
w_, v_ = power_iteration(np.dot(A.T, A))
print(w_.T[0], w)
print(np.round(v_.reshape(v.shape)))
print(v)
```
